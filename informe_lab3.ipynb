Análisis de la Gráfica

    Tendencia General:
        La gráfica muestra que a medida que se incrementa el número de hilos (n_threads), el tiempo de ejecución disminuye significativamente. Esto es un comportamiento esperado en aplicaciones paralelas donde más hilos permiten que el trabajo se divida y se ejecute simultáneamente, reduciendo el tiempo total de ejecución.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Datos organizados en un DataFrame para facilidad de uso
data = {
    'iterations': list(range(1, 11)),
    'n_threads_1': [2215.476, 2249.509, 2269.347, 2287.082, 2299.493, 2354.175, 2307.981, 2428.943, 2274.860, 2293.081],
    'n_threads_2': [1359.557, 1346.552, 1318.315, 1555.709, 1480.156, 1520.984, 1365.650, 1356.255, 1708.891, 1452.893],
    'n_threads_4': [982.706, 834.001, 843.387, 953.598, 959.935, 737.546, 1012.800, 885.929, 853.837, 1071.471],
    'n_threads_8': [633.267, 573.052, 607.225, 651.885, 626.457, 591.732, 651.292, 604.478, 638.251, 605.814],
    'n_threads_16': [503.150, 493.842, 505.471, 456.996, 479.745, 460.854, 467.565, 456.157, 483.121, 465.712]
}

df = pd.DataFrame(data)

# Crear DataFrame para boxplot
df_melted = df.melt(id_vars=['iterations'], var_name='n_threads', value_name='execution_time')

# Crear histogramas
plt.figure(figsize=(14, 10))

for i, n_thread in enumerate(['n_threads_1', 'n_threads_2', 'n_threads_4', 'n_threads_8', 'n_threads_16']):
    plt.subplot(3, 2, i+1)
    sns.histplot(df[n_thread], kde=True)
    plt.xlabel('Tiempo de Ejecución (ms)')
    plt.ylabel('Frecuencia')
    plt.title(f'Histograma: {n_thread.replace("_", " ")}')

plt.tight_layout()
plt.show()
```

Conclusiones

    Reducción del Tiempo de Ejecución: A medida que se incrementa el número de hilos, el tiempo de ejecución disminuye notablemente. Esto es evidente al comparar los histogramas de 1 hilo con los de 16 hilos.
    Variabilidad del Rendimiento: Las configuraciones con menos hilos (1 y 2) muestran una mayor variabilidad en los tiempos de ejecución. En contraste, las configuraciones con más hilos (8 y 16) presentan menos variabilidad, lo que indica un rendimiento más estable y consistente.
